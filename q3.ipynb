```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data.csv')
microscopy_df=df[df['sbelig']=='yes, selected for microscopy'] #checking how many were selected for microscopy
microscopy_df
```

```python
positive_malaria = (microscopy_df['hml32'] == 'positive').sum()
negative_malaria = (microscopy_df['hml32'] == 'negative').sum()
```

```python
total_tested=(microscopy_df['sbelig'].count())
```

```python
print(f' {total_tested} were selected for microscope test, {positive_malaria} are positive and {negative_malaria} are negative for malaria, while {total_tested-(positive_malaria+negative_malaria)} had blank results')
```
